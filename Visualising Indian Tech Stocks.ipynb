```python
import pandas as pd 
import numpy as np 
```

```python
import pandas_datareader as web
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
symbols = ['TCS', 'INFY', 'WIT']

start = '2020-01-01'
end = '2021-01-01'

stock_data = web.get_data_yahoo(symbols, start, end)

stock_data.head()
```

```python
stock_data['Adj Close'].head()
```

```python
stock_data['Adj Close'].plot()
plt.title("Tech Stocks Adjusted Price")
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price Over Time")
```

```python
daily_return = stock_data['Adj Close'].pct_change()
daily_return.plot()
plt.title("Daily Simple Rate of Return")
plt.xlabel('Date')
plt.ylabel("Rate of Return")
plt.show()
```

```python
print(daily_return.head())
```

```python
fig = plt.figure(figsize=(20, 10))


#TCS
ax1 = plt.subplot(2, 3, 1)
plt.plot(daily_return['TCS'], color='green')
plt.title('Tata Consultancy Services')
plt.xlabel('Date')
plt.ylabel('Daily Return')

#Infosys
ax2 = plt.subplot(2, 3, 2)
plt.plot(daily_return['INFY'], color='green')
plt.title('Infosys')
plt.xlabel('Date')


#Wipro Ltd.
ax3 = plt.subplot(2, 3, 3)
plt.plot(daily_return['WIT'], color='green')
plt.title('Wipro')
plt.xlabel('Date')
```

```python
mean_daily_return = daily_return.mean()
print(mean_daily_return)
```

```python
bar_chart = plt.subplot()
bar_chart.set_xticks(range(len(symbols)))
bar_chart.set_xticklabels(symbols)

plt.bar(range(len(symbols)), mean_daily_return, color='green')

plt.xlabel('Stocks')
plt.ylabel('Mean Daily')
plt.title("Mean Rate Of Return x Stocks")
```

```python
var = daily_return.var()
print(var)
```

```python
bar_chart = plt.subplot()
bar_chart.set_xticks(range(len(symbols)))
bar_chart.set_xticklabels(symbols)

plt.bar(range(len(symbols)), var, color='green')

plt.xlabel('Stocks')
plt.ylabel('Variance')
plt.title("Variance x Stocks")


```

```python
sd = daily_return.std()
print(sd)
```

```python
bar_chart = plt.subplot()
bar_chart.set_xticks(range(len(symbols)))
bar_chart.set_xticklabels(symbols)

plt.bar(range(len(symbols)), sd, color='green')

plt.xlabel('Stocks')
plt.ylabel('Standard Deiviation')
plt.title("Standard Deviation x Stocks")
```

```python
correlations = daily_return.corr()
print(correlations)
```
